```python
import importlib
import tiktoken

print("tiktoken version:", importlib.metadata.version("tiktoken"))
```

```python
tokenizer = tiktoken.get_encoding("gpt2")
```

```python
text = (
    "Hey there!<|endoftext|>My name is Tobias Rieper.<|endoftext|>"
     "I live in Madagascar."
)

integers = tokenizer.encode(text, allowed_special={"<|endoftext|>"})

print(integers)
```

```python
strings = tokenizer.decode(integers)

print(strings)
```

```python
integers = tokenizer.encode("Akwirw  idqloe ier GGWP english")
print(integers)

strings = tokenizer.decode(integers)
print(strings)
```
