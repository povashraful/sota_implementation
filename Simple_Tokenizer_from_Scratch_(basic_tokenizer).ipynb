```python
class llm_english_tokenizer:
    def __init__(self, vocab): # vocab is mapping tokens into token ids
        self.str_to_int = vocab # now the vocab will contain token and token IDs in this format string: int
        self.int_to_str = {i:s for s,i in vocab.items()} # we will take the strings and the token IDs and reverse them to this format (from string:int to int:string)

    def encode(self, text):
        preprocessed = re.split(r'([,.:;?_!"()\']|--|\s)', text)

        preprocessed = [
            item.strip() for item in preprocessed if item.strip()
        ]
        ids = [self.str_to_int[s] for s in preprocessed]
        return ids



    def decode(self, ids):
        text = " ".join([self.int_to_str[i] for i in ids]) # converting IDs into individual tokens and join the indiviual tokens together


        # Removing spaces before the specified punctuations
        # for example, 'Is the cat going to sleep  ?'
        # now we need to get this rid of the space before the puntuation mark "?"
        text = re.sub(r'\s+([,.?!"()\'])', r'\1', text)
        return text

```

```python
class llm_english_v2:
    def __init__(self, vocab):
        self.str_to_int = vocab
        self.int_to_str = { i:s for s,i in vocab.items()}

    def encode(self, text):
        preprocessed1 = re.split(r'([,.:;?_!"()\']|--|\s)', text)
        preprocessed1 = [item.strip() for item in preprocessed1 if item.strip()]
        preprocessed1 = [
            item if item in self.str_to_int
            else "<|unk|>" for item in preprocessed1 # if an item / word is not present in the vocabulary, then it should show unknown token.
        ]

        ids = [self.str_to_int[s] for s in preprocessed1]
        return ids

    def decode(self, ids):
        text = " ".join([self.int_to_str[i] for i in ids])
        # Replace spaces before the specified punctuations
        text = re.sub(r'\s+([,.:;?!"()\'])', r'\1', text)
        return text
```

```python
tokenizer = llm_english_v2(vocab)

text1 = "Hello there! Welcome to MIT! Kindly give me your ID."
text2 = "Hi! My name is El Professor. I was invited here as a guest lecturer. Here is my invitation and my ID."
text3 = "Thank you. Let me run the ID with our system. Ok, verified. Thank you for cooperating. Please now enter through the gate and the main hall will be on the right side after the entrance. Good Luck!"
text4 = "Thank you so much. I'll head out now."
text5 = "You are welcome. Have a blast."
text_random = "Hello, do you like tea?"



text = " <|endoftext|> ".join((text1, text2, text3, text4, text5, text_random))

print(text)
```

```python
# the " <|endoftext|> " is added betweeen the texts. They act as markers, telling us the start of end of a particular segment
```

```python
# This leads  to more effective processing and understanding by the llm.
```

```python
tokenizer.encode(text)
```

```python
tokenizer.decode(tokenizer.encode(text))
```

```python
class llm_simple_tokenizer_bangla_version:
    def __init__(self, vocab):
        self.str_to_int = vocab
        self.int_to_str = {i:s for s,i in vocab.items()}

    def encode(self, text):
        preprocessed = re.split(r'([,.:;?_!"()\']|--|\s)', text)

        preprocessed = [
            item.strip() for item in preprocessed if item.strip()
        ]
        ids = [self.str_to_int[s] for s in preprocessed]
        return ids

    def decode(self, ids):
        text = " ".join([self.int_to_str[i] for i in ids])
        # Replace spaces before the specified punctuations
        text = re.sub(r'\s+([,.?!"()\'])', r'\1', text)
        return text
```

```python
class llm_bangla_tokenizer_v2:
    def __init__(self, vocab4):
        self.str_to_int = vocab4
        self.int_to_str = { i:s for s,i in vocab4.items()}

    def encode(self, text):
        preprocessed4 = re.split(r'([,.:;?_!"()\']|--|\s)', text)
        preprocessed4 = [item.strip() for item in preprocessed4 if item.strip()]
        preprocessed4 = [
            item if item in self.str_to_int
            else "<|unk|>" for item in preprocessed4
        ]

        ids = [self.str_to_int[s] for s in preprocessed4]
        return ids

    def decode(self, ids):
        text = " ".join([self.int_to_str[i] for i in ids])
        # Replace spaces before the specified punctuations
        text = re.sub(r'\s+([,.:;?!"()\'])', r'\1', text)
        return text
```
